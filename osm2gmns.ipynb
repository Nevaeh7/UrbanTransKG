```python
import pandas as pd
import uuid
import hashlib
import pymongo
client = pymongo.MongoClient(host='127.0.0.1', port=27017)
md5 = hashlib.md5()
db = client['traffic']
node = db['gmns_nodes']
link = db['gmns_links']
df = pd.read_csv('node.csv',encoding='gbk')
node_df = df.dropna(subset=['name'])
node_df = node_df.fillna('')

node_dict = node_df.to_dict(orient='record')
for item in node_dict:
    md5.update((str(item['node_id']) + item['name']).encode('utf-8'))
    item['_id'] = md5.hexdigest()
    node.update({'_id':item['_id']},item,upsert=True)
    
df = pd.read_csv('link.csv',encoding='gbk')
node_df = df.dropna(subset=['name'])
node_df = node_df.fillna('')

node_dict = node_df.to_dict(orient='record')
for item in node_dict:
    md5.update((str(item['link_id']) + item['name']).encode('utf-8'))
    item['_id'] = md5.hexdigest()
    link.update({'_id':item['_id']},item,upsert=True)
    
```
